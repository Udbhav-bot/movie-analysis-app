```python
def fix_gross_val(value):
    if isinstance(value, str):
        value = value.replace(',','')
        return value
    return np.nan
movies['Gross'] = movies['Gross'].apply(fix_gross_val).astype(float)
```

```python
def fix_runtime_values(value):
    if isinstance(value, str):
        if 'min' in value:
            value = value.split()[0]
            return value
    return value

movies['Runtime'] = movies['Runtime'].apply(fix_runtime_values).astype(int)
movies['Runtime']
```

```python
movies.describe()
```

```python
movies.columns
```

```python
longest_runtime=movies[movies['Runtime']>180 ][['Series_Title','Runtime']].sort_values('Runtime',ascending=False)
longest_runtime
```

```python
def rating(rating):
    if rating>=9:
        return 'Excellent'
    elif rating>=8:
        return 'Good'
    elif rating>=7:
        return 'Nice'
    else:
        return 'Average'
```

```python
remark=movies['IMDB_Rating'].apply(rating)
```
